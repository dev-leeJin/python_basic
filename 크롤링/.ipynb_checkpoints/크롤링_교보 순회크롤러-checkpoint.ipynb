```python
# 크롤링 필요 라이브러리 임포트
from bs4 import BeautifulSoup # 웹에서 가져온 데이터를 정제하는 역할 (정돈)
import requests # 파이썬에서 다이렉트로 자료를 요청하는 역할
from selenium import webdriver # 크롬창을 켜주는 역할
import time # 코드 중간에 딜레이를 넣는 역할 (바로 실행이 되는걸 막고 딜레이를 넣어줌 )
```

```python
# 셀레니움 기본세팅을 해주세요.
# 알라딘 베스트셀러 페이지에 접속하도록 만들어주시면 됩니다.

browser = webdriver.Chrome('chromedriver.exe')
#browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1" % i)

# #순회크롤러를 해야하는데 주소창에 파라미터를 보면 page가 있는걸 볼 수 있는데 %s를 넣어줘서 페이지 이동이 가능하게 한다.
for i in range(1, 21):
    browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1" % i)
    time.sleep(2) # 2초의 텀을 주고 페이지가 1씩 증가하도록 함. 
```

```python
# 1페이지만을 대상으로 div.ss_bool_box를 50개가 잘 찍히는지 len을 이용해 찍어보기. 

browser = webdriver.Chrome('chromedriver.exe')
time.sleep(1)
browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=1&cnt=1000&SortOrder=1" )

source = browser.page_source
parsed_source = BeautifulSoup(source, "html.parser")

div_list = parsed_source.find_all("div", class_="ss_book_box")
print(len(div_list))
```

```python
# 가져온 50개에서 다시 개별 건에 대해 li를 추출해주신 다음
# 몇 번째 li가 제목인지
# 몇 번째 li가 가격인지 등을 파악
for item in div_list: # div_list는 책 정보가 50개까지 있음.,
   # for li in item.find_all("li"):
   #     print(li.text)
   # print("---------------------------------")
    li_list = item.find_all("li")
    print(li_list[1].text)
    print(li_list[2].text)
    print(li_list[3].text)
    print("------------")
    
```

```python
# #위에 이벤트 문구가 있고 없고 차이로 인덱스 번호가 다르게 나옴. -> len을 찍어서 있는거와 없는거를 구분
for item in div_list:
    li_list=item.find_all("li")
    li_len = (len(li_list))
    print(li_len)
```

```python
if (li_len < 8):
    for item in div_list:
        li_list = item.find_all("li")
        print(li_list[0].text)
        print(li_list[1].text)
        print(li_list[2].text)
        print("------------")
else:
    for item in div_list:
        li_list = item.find_all("li")
        print(li_list[1].text)
        print(li_list[2].text)
        print(li_list[3].text)
        print("------------")
        
# 실패
```

```python
# 해당 순위를 csv로 뽑기  #정답

for item in div_list :
    li_list = item.find_all("li")
    if str(li_list[0].text)[0] == "[": # 인덱스 0번에 대괄호 여는 부분이 있다면 1,2,3로 시작
        print(li_list[1].text)
        print(li_list[2].text)
        print(li_list[3].text)
        print("------------")
    else: # 없다면 0,1,2로 시작
        print(li_list[0].text)
        print(li_list[1].text)
        print(li_list[2].text)
        print("------------")
```

```python
# 페이지마다 다 뽑아내기. 
browser = webdriver.Chrome('chromedriver.exe')

for i in range(1, 21):
    browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1" % i)
    for item in div_list :
        li_list = item.find_all("li")
        if str(li_list[0].text)[0] == "[":
            print(li_list[1].text)
            print(li_list[2].text)
            print(li_list[3].text)
            print("------------")
        else: 
            print(li_list[0].text)
            print(li_list[1].text)
            print(li_list[2].text)
            print("------------")

    time.sleep(2)
```

```python
# csv로 뽑기

import codecs

f = codecs.open("C:/crawler1/aladin.txt", mode="w", encoding='utf-8')

browser = webdriver.Chrome('chromedriver.exe')

f.write("책제목,책저자,책가격\n")
for i in range(1, 21):
    browser.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BestType=Bestseller&BranchType=1&CID=0&page=%s&cnt=1000&SortOrder=1" % i)
    for item in div_list :
        li_list = item.find_all("li")
        if str(li_list[0].text)[0] == "[":
            f.write(li_list[1].text.replace(",","") + ",")
            f.write(li_list[2].text.replace(",","")  + ",")
            f.write(li_list[3].text.replace(",","")  + "\n")
        else: 
            f.write(li_list[0].text.replace(",","")  + ",")
            f.write(li_list[1].text.replace(",","")  + ",")
            f.write(li_list[2].text.replace(",","")  + "\n")

    time.sleep(2)
    
f.close()    

```
